```python
import pandas as pd
import numpy as np
from numpy import loadtxt

data = pd.read_csv (r'C:\Users\EWURAMA\Desktop\assignment\MScFE 650 MLF GWP Data.csv')
means = loadtxt(r'C:\Users\EWURAMA\Desktop\assignment\log_returns_rank.txt', delimiter=',')
vols = loadtxt(r'C:\Users\EWURAMA\Desktop\assignment\vol_rank.txt', delimiter=',')
print('\nmeans', means)
print('\nvols\n',vols)
data.head()
```

# 5.1 Calculate the Euclidean distance matrix for all 35 series

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

log_returns = data.select_dtypes(include=['float']).apply(np.log).diff().dropna()

distance_one = euclidean_distances(log_returns.to_numpy().T,log_returns.to_numpy().T)
```

```python
distance_one
```

```python
import seaborn as sns
distance_one = pd.DataFrame(distance_one, columns=data.columns[1:], index=data.columns[1:])
sns.heatmap(distance_one)
```

# Run a k-means cluster on the 35 return series and Justify your choice of k

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

result = []

for k in range(1,10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(log_returns.to_numpy())
    result.append(kmeans.inertia_)
    

plt.plot(range(1,10), result)
plt.xlabel('k')
```

```python
# Normalize each data series. Subtract each series by the weighted average from part 2.4.  
# Then divide this difference by the weighted volatility from part 3.4*
```

```python
#print("\nlog_returns:\n",log_returns) 
norm = (log_returns - means) / vols
norm.head()
```

```python
# Using the normalized data series, re-calculate the Euclidean distance matrix for all 35 series
```

```python
distance_two = euclidean_distances(norm.to_numpy().T, norm.to_numpy().T)
distance_two
```

```python
distance_two = pd.DataFrame(distance_two, columns=data.columns[1:], index=data.columns[1:])

sns.heatmap(distance_two)
```
