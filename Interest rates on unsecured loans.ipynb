```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter, YearLocator

# Load the data
file_path = "/content/TERMCBPER24NS.csv"
data = pd.read_csv(file_path)

# Convert 'DATE' column to datetime format
data['DATE'] = pd.to_datetime(data['DATE'])

# Explore basic characteristics of the data
print("First rows of the data:", data.head())
print("Shape of the data:", data.shape)
print("\nColumn names:\n", data.columns)
print("\nData types:\n", data.dtypes)

# Check for missing values
print("\nMissing values:\n", data.isnull().sum())

# Summarize the data
print("\nSummary statistics:\n", data.describe())

# Visualize the data
# Example: Line plot of the data column over time
plt.figure(figsize=(12, 8))  # Adjust figure size for better readability
plt.plot(data['DATE'], data['TERMCBPER24NS'])
plt.title('Finance Rate on Personal Loans at Commercial Banks (24-Month Loan)')
plt.xlabel('Date')
plt.ylabel('Rate (%)')

# Set date ticks with a 5-year difference
plt.gca().xaxis.set_major_locator(YearLocator(base=5))

# Format the date ticks as year
date_format = DateFormatter("%Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.xticks(rotation=45)  # Rotate the date labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()

# Example: Histogram of the data column
plt.figure(figsize=(10, 6))
sns.histplot(data['TERMCBPER24NS'], bins=20, kde=True)
plt.title('Histogram of Finance Rate on Personal Loans')
plt.xlabel('Rate (%)')
plt.ylabel('Frequency')
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
